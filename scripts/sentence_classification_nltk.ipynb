```python
def generate_pos_tags(text_string):
    """returns a list of tupples with pos tag attached with each word"""
    tokenizer = nltk.tokenize.TreebankWordTokenizer()
    word_list = tokenizer.tokenize(text_string)
    #print(word_list)
    pos_tag_list = nltk.pos_tag(word_list)
    #print(pos_tag_list)
    return pos_tag_list
```

```python
def generate_tag_list(tagged_post):
    """returns list of tags input= list of tupples, where second element = tag, first element = word"""
    tag_list = []
    word_list = []
    for mytup in tagged_post:
        tag_list.append(mytup[1])
        word_list.append(mytup[0])
    return tag_list, word_list
```

```python
def generate_ngrams(word_list, n):
    nltk_ngrams = nltk.ngrams(word_list, n)
    print(nltk_ngrams)
    ngrams_list = list(nltk_ngrams)
    new_list = []
    for tupple in ngrams_list:
        new_list.append(" ".join(tupple))
    # print(new_list)
    return new_list
```

```python
def dialogue_act_features_0(post):
    features= {}
    tokenizer = nltk.tokenize.TreebankWordTokenizer()
    word_list = tokenizer.tokenize(post)
    global block_count
    for word in words_list: #nltk.word_tokenize(post):
        tmp = word.lower()
        if tmp == "blocking" or tmp == "block" or tmp =="blocked" or tmp == "blocks":
            block_count += 1
        features['contains({})'.format(word.lower())] = True
    return features
```

```python
def dialogue_act_features_1(post):
    features= {}
    tokenizer = nltk.tokenize.TreebankWordTokenizer()
    word_list = tokenizer.tokenize(post)
    new_words_list = word_list + generate_ngrams(word_list,2) + generate_ngrams(word_list,3)
    for word in new_words_list: #nltk.word_tokenize(post):
        features['contains({})'.format(word.lower())] = True
    return features
```

```python
def dialogue_act_features_2(tagged_post):
    """generate features, input = list of tupples, first = word, second = tags of a post"""
    # print(tagged_post)
    features= {}
    tag_list = generate_tag_list(tagged_post)
    ngram_tag_list = tag_list + generate_ngrams(tag_list,2) + generate_ngrams(tag_list,3)
    for word in ngram_tag_list: #nltk.word_tokenize(post):
        features['contains({})'.format(word.lower())] = True
    return features
```

```python
def check_sentence_polarity(sentence):
    """check if some sentence is negative or positive"""
    word_list = sentence.split()
    print(word_list)
    word_set = set(word_list)
    negation = 0 # 0 ==> not negative intent, 1 ==> negative intent
    for phrase in negation_list:
        if phrase in word_set:
            negation = 1
            break
    return negation
```

```python
def expand_contractions(sentence):
    """returns the sentence with contractions expanded such as cann't ==> can not"""
    word_list = sentence.split()
    for i in range(len(word_list)):
        if word_list[i] in contractions.keys():
            word_list[i] = contractions[word_list[i]]
        if word_list[i] == 'i':
            word_list[i] = 'I'
    sentence = " ".join(word_list)
    return sentence
```
