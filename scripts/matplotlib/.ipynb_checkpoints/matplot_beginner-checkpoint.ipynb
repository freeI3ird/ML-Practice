```python
names = ['a', 'b', 'c']
values = [1, 10, 100]
plt.figure(1, figsize = (9,3))  # creates a figure
plt.subplot(1, 3, 1)      # creates subplot inside figure
plt.bar(names, values)    # creates bar plot inside above subplot
plt.subplot(1, 3, 2)      # creates another subplot
plt.scatter(names, values)# creates scatter plot
plt.subplot(1, 3, 3)      # creates another subplot
plt.plot(names, values)   # creates a plot with lines and markers.
plt.suptitle('Categorical plotting')
plt.show()
#plt.show()
```
