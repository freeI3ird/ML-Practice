# Color Model

### HSV
1. Hue Saturation Value 
    - Represented by cylinder 
    - Hue: Defines the color. 
        - Represented by angle of cylinder 
        - Range 0 to 360
    - Saturation: Defines grayscale or color
        - Range 0 to 100 or 0 to 1 
        - 0 means no color, grayscale. 
        - Represented by outerness of cylinder. At centre Saturation = 0
    - Value: Defines the intensity or brightness of color 
        - Range 0 to 100 or 0 to 1
        - 0 means black color 
        - Represented by the height of cylinder

## Subplots and Zorder 
1. Zorder: In a single figure, if we are drawing 2 or more things, zorder defines which drawing is first created. 
    - Lower zorder drawing will be created first.

```python
np.random.seed(34)
x = np.random.random(15)
y = np.random.random(15)
x
```

```python
# Zorder 
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,y, 'r', linewidth=3)
plt.scatter(x,y, s=120)
plt.title('Lines on the top of dots')

plt.subplot(2,1,2)
plt.plot(x,y,'r', linewidth=3, zorder=0)
plt.scatter(x,y, s=120, zorder =1)
plt.title('Dots on the top of lines')
plt.tight_layout()
```

### Origin and Extent 
1. Extent: Defines the Bounding box of data coordinates
    - Image will fill this bounding box
2. Origin: How data/image is filled in the bounding box
3. Scenaios: 
    1. Extent=None, Origin='upper'
        - BB = (-0.5, numcols-0.5, numrows-0.5, -0.5)
        - (0,0) will be at top left
    2. Extent = None, Origin = 'lower'
        - BB = (-0.5, numcols-0.5, -0.5, numrows-0.5)
        - (0,0) will be at bottom left
    3. 

```python
BB = (-1,10,-1,10)
BB2 = (-1,10, 10, -1)
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)
plt.imshow(data, extent=BB2)
plt.xlim(-2,11)
plt.ylim(-2,11)
plt.title(f'extent = {BB2}')

plt.subplot(1,2,2)
plt.imshow(data, extent= BB2, origin='upper')
plt.xlim(-2,11)
plt.ylim(-1,10)
plt.title(f'extent = {BB2}')
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,8))
axs[0,0].imshow(data)

axs[0,1].imshow(data, origin='lower')

axs[1,0].imshow(data, extent=BB)

axs[1,1].imshow(data, extent=BB)
axs[1,1].set_xlim(-2,11)
axs[1,1].set_ylim(-2,11)
```
